```python
from atomcloud.functions import MultiFunction1D

data_length = 256
x = np.linspace(0, data_length - 1, data_length)
bimodal_equations = ['gaussian', 'parabola', 'foffset']
bimodal_func = MultiFunction1D(bimodal_equations)

gaussian_x0 = data_length * (1 / 4)
gaussian_std = 10
parabola_x0 = data_length * (3 / 4)
parabola_r = 15
offset = .3

bimodal_parameters = [[1.2, gaussian_std, gaussian_x0],
                      [1.5, parabola_r, parabola_x0],
                      [offset]]

bimodal_data = bimodal_func(x, bimodal_parameters)

plt.figure()
plt.plot(x, bimodal_data)
```

We can also do this for 2D functions. Below we use a fixed enhanced bose (fugacity=1), a thomas-fermi profile and a fixed offset

```python
from atomcloud.functions import MultiFunction2D
from atomcloud.utils import img_utils


coordinates = img_utils.get_coordinates(shape=(256, 256))

bimodal_equations = ['febose', 'tf', 'foffset']
bimodal_func = MultiFunction2D(bimodal_equations)


x0 = data_length / 2
y0 = x0

parabola_rx = 10.4
parabola_ry = 13.1
gaussian_stdx = 31.2
gaussian_stdy = 28
theta = np.pi / 3
offset = 0.2

bimodal_parameters = [[1.2, x0, y0, gaussian_stdx, gaussian_stdy, theta],
                      [1.5, x0, y0, parabola_rx, parabola_ry, theta],
                      [offset]]

bimodal_data = bimodal_func(coordinates, bimodal_parameters)

plt.figure()
plt.pcolormesh(*coordinates, bimodal_data)
plt.show()
```

```python
bimodal_func = MultiFunction2D(bimodal_equations, use_jax=True)

bimodal_data = bimodal_func(coordinates, bimodal_parameters)
print('bimodal_data data type:', type(bimodal_data))

plt.figure()
plt.pcolormesh(*coordinates, bimodal_data)
plt.show()
```

## Defining Custom Sub-Functions

By default, only 'gaussian', 'parabola', 'tf', 'ebose' 'febose' and 'foffset' are defined. However, this is a fairly limited number of functions and we might wish to add our own. All the information about the functions is defined in the 1d and 2d function registries. Each of these contains 1d and 2d fit object. These fit objects all inherit from the common FunctionBase class which looks like this

```python
class FunctionBase(ABC):
    """Base class for function objects"""

    def __init__(self):
        """Instantiates the function object, the only argument is whether
        to use jax for the function"""
        self.function = self.make_function(use_jax=False)
        self.create_parameter_dict()

    def __call__(self,
                 coords: Union[np.ndarray, Iterable[np.ndarray]],
                 *params: float
                 ) -> np.ndarray:
        return self.function(coords, *params)

    def make_function(self, use_jax: bool = False) -> callable:
        if use_jax:
            if jnp is not None:
                return self.create_function(jnp)
            else:
                raise Exception('JAX is not installed')
        else:
            return self.create_function(np)

    def create_parameter_dict(self) -> None:
        """Creates a dictionary of the parameters of the function"""
        self.param_dict = list(signature(self.function).parameters)[1:]

    def analyze_parameters(self, params: list[float]) -> dict:
        return {}

    def rescale_parameters(self,
                           params: list[float],
                           scales: list
                           ) -> list[float]:
        return params

    def rescale_analysis_params(self,
                                params: dict,
                                scales: list) -> dict:
        return params

    def initial_seed(self,
                     coords: Union[np.ndarray, Iterable[np.ndarray]],
                     data: np.ndarray) -> list[float]:
        return [1.0 for _ in self.param_dict]

    def default_bounds(self) -> tuple[list[float], ...]:
        min_bounds = [-np.inf for _ in self.param_dict]
        max_bounds = [np.inf for _ in self.param_dict]
        return (min_bounds, max_bounds)
```

The various functions such as initial seed and default bounds have to do with fitting, whereas things like analyze parameters allows post processing to be done when the function is used in a fit. See processing fits for more information.


Now let's inherit this object, make our own fit object and then add it to the packages registry

```python
from atomcloud.functions import FunctionBase, FUNCTIONS1D
from atomcloud.analysis import rescale_1d_params


class Cos1D(FunctionBase):
    """See FunctionBase for documentation"""

    def __init__(self):
        super().__init__()
        
    def create_function(self, anp):
        def exp_func(x, a, b, x0):
            return a * anp.cos(b * (x - x0))
        return exp_func

    def initial_seed(self, x, data):
        min_ind = np.argmin(data)
        max_ind = np.argmax(data)
        a = (data[max_ind] - data[min_ind]) / 2
        xdelta = x[-1] - x[0]
        x0 = xdelta / 2
        b = xdelta / 12
        return [a, std, x0]

    def rescale_parameters(self, params, scale):
        _, axis_scale, zscale = scale
        zinds = [0]
        xinds = [1, 2]
        return rescale_1d_params(params, zinds, xinds, axis_scale, zscale)

    def default_bounds(self):
        min_bounds = [0, 0, -np.inf]
        max_bounds = [np.inf, np.inf, np.inf]
        return [min_bounds, max_bounds]
    
FUNCTIONS1D.register('cos', Cos1D)
```

```python
from atomcloud.functions import MultiFunction1D

data_length = 256
x = np.linspace(0, data_length - 1, data_length)
cos_equations = ['gaussian', 'cos', 'foffset']
cos_func = MultiFunction1D(cos_equations)

gaussian_amp = 1.2
gaussian_x0 = data_length * (1 / 4)
gaussian_std = 10
cos_x0 = data_length * (3 / 4)
cos_amp = .5
cos_freq = 1
offset = .3

cos_parameters = [[gaussian_amp, gaussian_std, gaussian_x0],
                  [cos_amp, cos_freq, cos_x0],
                  [offset]]

cos_data = bimodal_func(x, cos_parameters)

plt.figure()
plt.plot(x, cos_data)
plt.show()
```

Or how about adding a 2D equation

```python
from atomcloud.functions import FunctionBase, FUNCTIONS2D
from atomcloud.analysis import rescale_2d_params

class Sin2D(FunctionBase):
    """See FunctionBase for documentation"""

    def __init__(self):
        super().__init__()
        
    def create_function(self, anp):
        def a_sin_func(coords, a, b, c):
            x, y = coords
            return a * np.sin(b * x + c * y) 
        return a_sin_func

    def initial_seed(self, x, data):
        min_ind = np.argmin(data)
        max_ind = np.argmax(data)
        a = (data[max_ind] - data[min_ind]) / 2
        xdelta = x[-1] - x[0]
        b = xdelta / 10
        c = xdelta / 12
        return [a, std, x0]

    def rescale_parameters(self, params, scale):
        _, axis_scale, zscale = scale
        zinds = [0]
        xinds = [1, 2]
        return rescale_2d_params(params, zinds, xinds, axis_scale, zscale)

    def default_bounds(self):
        min_bounds = [0, 0, -np.inf]
        max_bounds = [np.inf, np.inf, np.inf]
        return [min_bounds, max_bounds]
    
FUNCTIONS2D.register('sin', Sin2D)
```

Now let's try it

```python
from atomcloud.functions import MultiFunction2D
from atomcloud.utils import img_utils


coordinates = img_utils.get_coordinates(shape=(256, 256))

sin_equations = ['gaussian', 'sin', 'foffset']
sin_func = MultiFunction2D(sin_equations)


x0 = data_length / 2
y0 = x0

sin_freq_x = .1
sin_freq_y = .2
gaussian_stdx = 31.2
gaussian_stdy = 28
theta = np.pi / 3
offset = 0.2

sin_parameters = [[1.2, x0, y0, gaussian_stdx, gaussian_stdy, theta],
                  [1.5, sin_freq_x, sin_freq_y],
                  [offset]]

sin_data = sin_func(coordinates, sin_parameters)

plt.figure()
plt.pcolormesh(*coordinates, sin_data)
plt.show()
```
